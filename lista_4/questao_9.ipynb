# Imports necess√°rios

```python
import numpy as np

import matplotlib.pyplot as plt

import sympy as sp
```

# Letra A

## Calculando coeficientes

```python
h = [1, 2, 3, 4, 5, 6, 7, 8]

c = [8, 12.3, 15.5, 16.8, 17.1, 15.8, 15.2, 14.0]

fi0 = []

fi1 = []

fi2 = []

F = []

# Preenche matrizes
for i in range(len(h)):
    fi0.append(1)
    fi1.append(np.log(h[i]))
    fi2.append(h[i])
    F.append(np.log(c[i]))


A = [
    [np.dot(fi0, fi0), np.dot(fi0, fi1), np.dot(fi0, fi2)],
    [np.dot(fi1, fi0), np.dot(fi1, fi1), np.dot(fi1, fi2)],
    [np.dot(fi2, fi0), np.dot(fi2, fi1), np.dot(fi2, fi2)],
]

b = [[np.dot(F, fi0)], [np.dot(F, fi1)], [np.dot(F, fi2)]]

# Coeficientes

coef = np.linalg.solve(A, b)

print("coef = \n ", coef)

c1 = np.exp(coef[0])
c2 = coef[2]

print("\n\n C1=", c1[0], "e C2=", c2[0])
```

## Plotando resultados

```python
fig = plt.figure(figsize=[16, 9])
fig.suptitle("Concentra√ß√£o em fun√ß√£o do tempo", fontsize=16)

# Plotango 3D
ax = fig.add_subplot(1, 1, 1)

x_eval = np.linspace(0, 10, num=100)
c = [c1[0] * x * sp.exp(c2[0] * x) for x in x_eval]

ax.plot(x_eval, c)

ax.set_ylabel("Concentra√ß√£o", fontsize=16)
ax.set_xlabel("$t$", fontsize=16)

# Configurar a grade no gr√°fico
plt.grid()

# Mostrar o gr√°fico
plt.show()

# Imprimir a fun√ß√£o polinomial g
```

# Letra B

```python
t = sp.symbols("t")
csimb = c1[0] * t * sp.exp(c2[0] * t)

csimb_derivada = sp.diff(csimb, t)

ponto_critico = sp.solve(csimb_derivada, t)

max_valor = max([csimb.subs(t, cp) for cp in ponto_critico])

print(
    "A concentra√ß√£o m√°xima √© de {:.6f} e ocorre quando j√° se passaram {:.6f}".format(
        max_valor, ponto_critico[0]
    ),
    "horas.",
)
```
